```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

columns = ["L", "ops/sec"]
df = pd.read_csv("data.csv", usecols=columns)

x = df['L'].to_numpy()
y = df['ops/sec'].to_numpy()

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("L")
plt.ylabel("ops/sec")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("data1millwithbulk.csv")
print(df)
import matplotlib.pyplot as plt
import numpy as np
plot_str = 'loadtime'
k = df['K'].to_numpy()
l = df['L'].to_numpy()
c = df[plot_str].to_numpy()
#x = [np.sqrt(k[i]*k[i] + l[i]*l[i]) for i in range(len(df['K']))]
#y = df['ops/sec'].to_numpy()
# plt.scatter(k, l, c=c, cmap='viridis')
c = c.reshape(11, -1).transpose()
plt.imshow(c, origin='lower', cmap='viridis')
plt.colorbar(label=plot_str)
plt.xlabel("K")
plt.ylabel("L")
plt.xticks(range(11), labels=list(range(0, 101, 10)))
plt.yticks(range(11), labels=list(range(0, 101, 10)))
plt.title("Bulk load")
plt.show()
```
