```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

columns = ["insert_frac", "ops/sec"]
df = pd.read_csv("data.csv", usecols=columns)

x = df['insert_frac'].to_numpy()
y = df['ops/sec'].to_numpy()

plt.plot(x,y)
plt.xlabel("insert_frac")
plt.ylabel("ops/sec")
plt.show()
```
