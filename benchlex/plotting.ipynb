```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

columns = ["L", "ops/sec"]
df = pd.read_csv("data.csv", usecols=columns)

x = df['L'].to_numpy()
y = df['ops/sec'].to_numpy()

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("L")
plt.ylabel("ops/sec")
plt.show()
```
